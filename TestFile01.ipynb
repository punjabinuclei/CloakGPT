```python
model = GLiNER.from_pretrained("urchade/gliner_multi_pii-v1")


# for entity in entities:
#     print(entity["text"], "=>", entity["label"])
```

```python
text = input("Enter your prompt: ")

labels = ["work","card number", "ID Number", "Web Page" "booking number","Number", "Person", "personally identifiable information", "driver licence", "name", "book", "full address", "company", "actor", "character", "email", "passport number", "Social Security Number", "phone number", "IP Address", "Website"]
result = model.predict_entities(text, labels)

print(result)
```

```python
def replace_pii(text, result):
    # Sort entities by start index in descending order to handle replacements from the end of the string
    result = sorted(result, key=lambda x: x['start'], reverse=True)

    for entity in result:
        placeholder = f"[{entity['label'].upper()}]"
        start = entity['start']
        end = entity['end']

        # Replace the PII in the text
        text = text[:start] + placeholder + text[end:]

    return text
```

```python
modified_text = replace_pii(text, result)
print(modified_text)
```

```python
# Groq
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Repeat whatever i write next after fullstop as i am trying to test something." + modified_text,
        }
    ],
    model="llama3-8b-8192",
)
```
